```python
import numpy as np
import matplotlib.pyplot as plt
import os

# -------------------------------------------------------------------
# Learning rates to load
# -------------------------------------------------------------------
learning_rates = [1e-1, 5e-2, 1e-2, 1e-3]

results_folder = "results"
results = {}
max_len = 250   # first 250 epochs only

# -------------------------------------------------------------------
# Load and normalize each NPZ file
# -------------------------------------------------------------------
for lr in learning_rates:
    fname = f"{results_folder}/Adam_lr{lr:.4g}.npz"

    if not os.path.exists(fname):
        print(f" Missing file: {fname} (skipping).")
        continue

    data = np.load(fname)
    losses = data["train_losses"][:max_len]

    max_loss = losses.max() if len(losses) > 0 else 1.0
    normalized = losses / max_loss

    results[lr] = normalized
    print(f"Loaded {fname}: using {len(normalized)} points, max={max_loss:.4f}")


# -------------------------------------------------------------------
# Plot normalized training-loss curves
# -------------------------------------------------------------------
plt.figure(figsize=(10, 6))

for lr, norm_losses in results.items():
    plt.plot(norm_losses, label=rf"$\eta = {lr:g}$", lw=2)

#plt.title("Normalized Training Loss (First 1000 Epochs) â€” Adam")
plt.xlabel("Epoch")
plt.ylabel("Normalized Loss (loss / max(loss))")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.tight_layout()

# Save normalized plot
output_file = f"{results_folder}/Adam_training_loss_first1000_normalized.png"
plt.savefig(output_file, dpi=300)
print(f"Saved plot -> {output_file}")

plt.show()
```
